#Data Cleaning

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#load the dataset
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
#drop blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for unrelated\null values
pd.isnull(df)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#change the data type of 'Amount' column to integer
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
#rename the column 'Shaadi' to 'Marital_Status'
df.rename(columns={'Shaadi':'Marital_Status'}, inplace=True)
```

```python
df.describe()
```

```python
df[['Age','Orders' ,'Amount']].describe()
```

EDA

```python
#countplot for Gender
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#barplot for Gender vs Amount
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
df.columns
```

```python
#countplot for Age Group
age=sns.countplot(x='Age Group',data=df,hue='Gender',palette='Set2')
sns.set(rc={'figure.figsize':(10,5)})

for bars in age.containers:
    age.bar_label(bars)
```

```python
#barplot for Age Group vs Amount
saleage=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Age Group',y='Amount',data=saleage)
```

```python
df.columns
```

```python
# total number of ordrs top 10 states
state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data=state)
```

```python
#total  amount top 10 states
state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Amount',data=state)
```

```python
df.columns
```

```python
#marital status 
status=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(5,5)})

for bars in status.containers:
    status.bar_label(bars)
```

```python
#marital status with gender
status=sns.countplot(x='Marital_Status',data=df,hue='Gender',palette='Set2')
sns.set(rc={'figure.figsize':(5,5)})

for bars in status.containers:
    status.bar_label(bars)
```

```python
#barplot for Marital Status vs Amount
sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales,hue='Gender')
```

```python
#occupation with amount
occ=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,10)})

for bars in occ.containers:
    occ.bar_label(bars)
```

```python
occ=sns.countplot(x='Occupation',data=df,hue='Gender',palette='Set2')
sns.set(rc={'figure.figsize':(20,10)})

for bars in occ.containers:
    occ.bar_label(bars)
```

```python
#occupation with amount
sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Occupation',y='Amount',data=sales)
```
